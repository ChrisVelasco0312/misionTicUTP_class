```python
# Libreria para graficar con datos
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

fig, ax = plt.subplots(ncols=3, figsize=(16,4))


data = pd.read_csv('devs.csv')
sns.pointplot(x='salario', y='years', data=data, ax=ax[0])
sns.pointplot(x='salario', y='gender', data=data, ax=ax[1])
sns.pointplot(x='gender', y='years', data=data, ax=ax[2])



```
